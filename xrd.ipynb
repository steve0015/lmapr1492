```python
#Imports utiles

from pymatgen.ext.matproj import MPRester
from pymatgen import Structure
from pymatgen.symmetry.analyzer import SpacegroupAnalyzer
from pymatgen.analysis.diffraction.core import AbstractDiffractionPatternCalculator, DiffractionPattern
from pymatgen.analysis.diffraction.xrd import XRDCalculator
from pymatgen.analysis.diffraction.tem import TEMCalculator
```

```python
#On récupère la structure

with MPRester("8M635zkNlb27ZqrS") as m:
    Structure = m.get_structure_by_material_id("mp-581024")
analyse_struct = SpacegroupAnalyzer(Structure)   
conventional_structure = analyse_struct.get_conventional_standard_structure()
```

```python
#On crée une instance de la classe XRDCalculator

XRD_calculator = XRDCalculator(wavelength='CuKa', symprec=0, debye_waller_factors=None)
```

```python
#On récupère les données pour le difractogramme

pattern = XRD_calculator.get_pattern(structure = Structure, scaled=True, two_theta_range=(0, 90))
```

```python
#On trace le diffractogramme 

print("Diffractogramme :")
XRD_calculator.show_plot(structure = conventional_structure, annotate_peaks = True)
```

```python
print("\n\nOn remarque sur le diagramme que l'intensité du premier pic est petite par rapport aux deuxième et troisième pic. Les intensités sur le diagramme sont représentés en relatif en considérant la plus grande intensité, donc du pic associé à hkl = (2,2,0). L'intensité de ce pic est alors de 100% et les autres intensités sont données en relatif par rapport à celle-ci. On peut alors conclure que l'intensité du premier pic a une intensité à peu près 600 fois plus petites que celle du  troisième pic. ")
```

```python
print("Informations dans le pattern :")
print(pattern)
```

```python
print("\n\nLe premier pic du diffractogramme a des indices hkl = (1,1,1).\nL'angle associé à ce pic est 2*Theta =  17.538°, donc Theta = 8.469°.")
print("L'intensité relative vaut", pattern.y[0], "(en pourcentages par rapport aux 100% du pic associé à hkl = (2,2,0)).")
print("La multiplicité est", pattern.hkls[0] [0] ['multiplicity'],".")
```
